```python
from sqlquerytools import SqlQueryResult
```

```python
sql_code = """SELECT TOP(10) *
FROM CLIENT AS C WITH (NOLOCK)
WHERE C.ARCHIVEPARTITION IN (0,1)
OPTION (MAXDOP 1);"""
sql_code2 = """SELECT 
	C.RESIDENTIALSTATECODE
	,COUNT(*) AS CLIENT_COUNT
FROM CLIENT C WITH (NOLOCK)
WHERE
  C.ARCHIVEPARTITION IN (0,1)
  AND C.ISTEST = 0
GROUP BY C.RESIDENTIALSTATECODE
Order by CLIENT_COUNT Desc
OPTION (MAXDOP 1)"""
```

```python
z_sql = SqlQueryResult()

# Query
z_sql.query(sql_code)

# Results return in Pandas DataFrame.
q1 = z_sql.result()
q2 = z_sql.result(sql_code2)
```

```python
display(q2)
```
